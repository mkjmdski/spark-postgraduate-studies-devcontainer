```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("spark-minio") 
    .enableHiveSupport()
    .getOrCreate()
)
```

```python
spark
```

```python
# creating new table in metastore

df = spark.range(10)
df.write.mode("overwrite").saveAsTable("test_minio_v1")
```

```python
spark.sql("select * from test_minio_v1").show()
```

```python
spark.sql("show create table test_minio_v1").show(truncate=False)
```

```python
spark.sql("drop table test_minio_v1")
```

```python
# loading data from s3 to a table

df = spark.read.format("csv").option("header", True).load("s3a://landing-zone/s3-sample.csv")
df.write.saveAsTable("test_minio_v2")
```

```python
spark.sql("select * from test_minio_v2").show()
```

```python
spark.sql("show create table test_minio_v2").show(truncate=False)
```

```python
spark.stop()
```
